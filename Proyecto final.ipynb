Pregunta 1: ¿Cuál es el porcentaje de transacciones fraudulentas en el dataset?



```python
# Calculamos el porcentaje de transacciones fraudulentas
def calculoporcentaje(dato, columna):
    contador_transacciones_fraudes = (dato[columna] == 1).sum()
    contador_total_transacciones = len(dato)

    if contador_total_transacciones > 0:
        porcentaje_fraude = (contador_transacciones_fraudes / contador_total_transacciones) * 100
    else:
        porcentaje_fraude = 0
        
    return porcentaje_fraude
        
# Mostramos el porcentaje de transacciones fraudulentas
porcentaje_fraude = calculoporcentaje (data,'Class')
print("El porcentaje de transacciones fraudulentas es: {:.2f}%".format(porcentaje_fraude))
```

Pregunta 2: ¿Cuál es el importe medio de las transacciones fraudulentas?



```python
# Calculamos el importe medio de las transacciones fraudulentas
transacciones_fraudulentas = data[data['Class'] == 1]
importe_medio = transacciones_fraudulentas['Amount'].mean()

# Mostramos el importe medio de las transacciones fraudulentas
print("El importe medio de las transacciones fraudulentas es: {:.2f}".format(importe_medio))
```


Pregunta 1: ¿Cuántas transacciones fraudulentas hay en comparación con las no fraudulentas? (Utiliza un gráfico de barras)

```python
# Contamos el número de transacciones fraudulentas y no fraudulentes
conteo = data['Class'].value_counts()

transacciones_fraudulentas = conteo.get(1, 0)  
transacciones_no_fraudulentas = conteo.get(0, 0) 

# Mostramos la distribución de las traducciones fraudulentas con respecto de las no fraudulentas
print("Las transacciones fraudulentas son: {:.2f}".format(transacciones_fraudulentas))
print("Las transacciones no fraudulentas son: {:.2f}".format(transacciones_no_fraudulentas))
```

Gráfico de las transacciones fraudulentas en comparación con las no fraudulentas

```python
diccionario = {'Class': conteo.index, 'Count': conteo.values}
distribucion_transacciones_fraudulentas = pd.DataFrame(diccionario)
sns.barplot(x = 'Class',y = 'Count',data = distribucion_transacciones_fraudulentas)
plt.xticks(ticks=[0, 1], labels=['No Fraude', 'Fraude'])
plt.xlabel('Tipo de Transacción')
plt.ylabel('Cantidad de Transacciones')
plt.title('Transacciones Fraudulentas vs No Fraudulentas')
plt.show()
```

Pregunta 2: ¿Cuál es la distribución de los importes de las transacciones fraudulentas? (Utiliza un histograma)


```python
# Separamos los datos de transacciones fraudulentas
transacciones_fraudulentas = data[data['Class'] == 1]

# Mostrando la distribución de los importes de las transacciones fraudulentas
plt.figure(figsize=(10, 6))
sns.histplot(transacciones_fraudulentas['Amount'], bins=50)
plt.xlabel('Importe de la Transacción')
plt.ylabel('Cantidad de Transacciones')
plt.title('Distribución de los importes de las Transacciones Fraudulentas')
plt.show()
```

## Desarrollo y evaluación de modelos

```python
from sklearn.model_selection import train_test_split
```

```python
# Separamos los datos de entrenamiento y evaluación
X = data.drop('Class',axis=1)
y = data['Class']
```

```python
X
```

```python
y
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size = 0.2, random_state = 45)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
```

```python

modelo = RandomForestClassifier(max_depth=150, random_state=42)
```

```python
modelo.fit(X_train, y_train)
```

```python
# Realizando predicciones en el conjunto de prueba
y_pred = modelo.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
# Evaluando el modelo
print(classification_report(y_test, y_pred))
```

```python
exactitud = accuracy_score(y_test,y_pred)
print("Exactitud del modelo: {:.2f}%".format(exactitud * 100))
```
